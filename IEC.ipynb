```python
import os
jpg=[]
jpeg=[]
png=[]
mix=[]
for root, dirs, files in os.walk(r"C:\Users\angshuman\Desktop\IEC\dest"):
    for i in files:
        if i.endswith(".jpg"):
            jpg.append(os.path.join(root, i))
        elif i.endswith(".png"):
            png.append(os.path.join(root, i))
        elif i.endswith(".jpeg"):
            jpeg.append(os.path.join(root, i))
        else:
            mix.append(os.path.join(root, i))
            
```

```python
jpg1=[]
for roott, dirrs, filess in os.walk(r"C:\Users\angshuman\Desktop\IEC\JPG"):
    for ii in filess:
        if ii.endswith(".jpg"):
            jpg1.append(os.path.join(roott, ii))
```

```python
jpeg1=[]
for roott1, dirrs1, filess1 in os.walk(r"C:\Users\angshuman\Desktop\IEC\JPEG"):
    for ii1 in filess1:
        if ii1.endswith(".jpeg"):
            jpeg1.append(os.path.join(roott1, ii1))
```

```python
mix1=[]
for roott2, dirrs2, filess2 in os.walk(r"C:\Users\angshuman\Desktop\IEC\MIX"):
    for ii2 in filess2:
        mix1.append(os.path.join(roott2, ii2))
```

### JPG TO PNG

```python
from PIL import Image
c=0
for i in range(len(jpg1)):
    try:
        im1 = Image.open(jpg[i])
        f=jpg[i]
        f=f.split('\\')
        s=f[6][:-4]
        im1.save("C:\\Users\\angshuman\\Desktop\\IEC\\JPG TO PNG\\"+s+".png")
    except:
        c=c+1
        print(jpg1[i])
        
print("Error files",c)
```
